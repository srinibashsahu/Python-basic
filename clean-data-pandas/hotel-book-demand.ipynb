```python
# create new data frame which will have catagoricat data
cat_cols= [col for col in df_data.columns if df_data[col].dtype== "object"]
cat_cols
```

```python
cat_df= df_data[cat_cols]
cat_df.head()
```

```python
# printing unique value 
for col in cat_df.columns:
    print(f"{col}:\n {cat_df[col].unique()}\n")
```

```python
# manually map arrival date month to int value
cat_df["arrival_date_month"] = cat_df["arrival_date_month"].map({'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12, 'January':1,
 'February':2, 'March':3, 'April':4, 'May':5, 'June':6})
cat_df.head()
```

```python
#label encode country and hotel
!pip install scikit-learn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_df["country"] =le.fit_transform(cat_df["country"])
cat_df["hotel"]= le.fit_transform(cat_df["hotel"])
cat_df.head()
```

```python
# printing unique value 
for col in cat_df.columns:
    print(f"{col}:\n {cat_df[col].unique()}\n")
```

```python
#one hot encoding 
cat_df = pd.get_dummies(data=cat_df, columns=["meal","market_segment","distribution_channel","reserved_room_type","assigned_room_type","deposit_type","customer_type","reservation_status"])
cat_df
```

```python
num_df = df_data.drop(columns=cat_cols,axis=1)
num_df
```

```python
final_df= pd.concat([num_df,cat_df],axis=1)
final_df
```

```python
#create a correlation matrix 
corr_matrix= final_df.corr().abs()
corr_matrix
```

```python
# setting to threshhold
threshold= 0.85
import numpy as np
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
#ensure only 1 of highly correlated feature get droped 
to_drop= []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i,j]) >= threshold:
            colname=corr_matrix.columns[i]
            to_drop.append(colname)

to_drop
```
