```python
import pandas as pd
column_name = [
    "Date of Transaction",
    "Branch Name",
    "Amount of Sales",
    "Is Active"
]
primary_data= {
    "Date of Transaction": ["Jan 5", "jan 6"],
    "Branch Name": ["Branch C", "Branch B"],
    "Amount of Sales":[200.00,400],
    "Is Active": [True,True]
}

df_primary= pd.DataFrame(primary_data,columns=column_name)
df_primary
```

```python
default_data= {
    "Date of Transaction": ["Jan 1", "jan 2"],
    "Branch Name": ["Branch A", "Branch B"],
    "Amount of Sales":[500.00,600],
    "Is Active": [True,False]
}
df_default=pd.DataFrame(default_data)
df_primary=pd.concat([df_primary,df_default])
df_primary.reset_index(drop=True,inplace=True)
df_primary
```
